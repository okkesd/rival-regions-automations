```python
import speedtest

def hiz_testi():
    st = speedtest.Speedtest()
    st.get_best_server()  # En iyi sunucuyu seç
    download_hiz = st.download() / 1_000_000  # Mbps'e çevir
    upload_hiz = st.upload() / 1_000_000  # Mbps'e çevir
    ping = st.results.ping  # Ping süresi (ms)

    print(f"İndirme Hızı: {download_hiz:.2f} Mbps")
    print(f"Yükleme Hızı: {upload_hiz:.2f} Mbps")
    print(f"Ping: {ping:.2f} ms")

hiz_testi()
```

```python
arr = [1,2,4,7,8,12]

i, j = 0, len(arr) -1

while i < j:
    print("i: " + str(arr[i]) + " , j: " + str(arr[j]))

    sum = arr[i] + arr[j]
    if (sum == 10):
        print("found: " + str(arr[i]) + " " + str(arr[j]))
        break
    elif sum > 10:
        j -= 1
    else:
        i += 1
    
```
