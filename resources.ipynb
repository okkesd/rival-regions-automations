```python
import cloudscraper
import zlib
from bs4 import BeautifulSoup
import time
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def printer(name):
    content_encoding = name.headers.get('Content-Encoding')

    # Decompress the response content based on the encoding
    if content_encoding == 'gzip':
        decompressed_content = zlib.decompress(name.content, zlib.MAX_WBITS | 16)
    elif content_encoding == 'deflate':
        decompressed_content = zlib.decompress(name.content)
    #elif content_encoding == 'br':
    #    decompressed_content = brotli.decompress(name.content)
    else:
        decompressed_content = name.content

    html_content = decompressed_content.decode('utf-8')

    soup = BeautifulSoup(html_content, 'html.parser')
    print(soup.prettify())
    
    
def souper(name):
    content_encoding = name.headers.get('Content-Encoding')

    # Decompress the response content based on the encoding
    if content_encoding == 'gzip':
        decompressed_content = zlib.decompress(name.content, zlib.MAX_WBITS | 16)
    elif content_encoding == 'deflate':
        decompressed_content = zlib.decompress(name.content)
    #elif content_encoding == 'br':
    #    decompressed_content = brotli.decompress(name.content)
    else:
        decompressed_content = name.content

    html_content = decompressed_content.decode('utf-8')

    soup = BeautifulSoup(html_content, 'html.parser')
    return soup
```

```python
# Drop duplicates
resources.drop_duplicates(subset=["State"], keep="first", inplace=True)
from datetime import datetime
time = str(datetime.now())[5:10]
month = time[0:2]
day = time[3:6]
with pd.ExcelWriter(f'resources_{month}_{day}.xlsx', engine='openpyxl') as writer:
    
    resources.to_excel(writer, sheet_name='Sheet1', index=True)
    # Openpyxl ile Excel dosyasını düzenleme
    workbook = writer.book
    worksheet = writer.sheets['Sheet1']
    
    # Sütun aralığını belirt
    column_range = f"A1:{chr(65+len(resources.columns))}{len(resources)+1}"
    print(column_range)
    # Tablo oluşturma
    worksheet.auto_filter.ref = column_range 
```

```python
blocs = {}
for name,group in resources.groupby("Bloc"):
    blocs[name] = group
#print(blocs.keys())
blocs_df = pd.DataFrame(columns=["Block", "Cash(T)", "Gold(T)", "Oil(kkk)", "Ore(kkk)", "Uranium(kk)", "Dia(kk)"])
for name in blocs.keys():
    cash = round(blocs[name]["Cash"].str.replace(".", "", regex=False).astype(float).sum() / 1000000000000,2)
    gold = round(blocs[name]["Gold"].str.replace(".","", regex=False).astype(float).sum() / 1000000000000,2)
    oil = round(blocs[name]["Oil"].str.replace(".","", regex=False).astype(float).sum() / 1000000000,2)
    ore = round(blocs[name]["Ore"].str.replace(".","", regex=False).astype(float).sum() / 1000000000,2)
    uranium = round(blocs[name]["Uranium"].str.replace(".","", regex=False).astype(float).sum() / 1000000 ,2)
    dia = round(blocs[name]["Dia"].str.replace(".","", regex=False).astype(float).sum() / 1000000 ,2)
    
    blocs_df = blocs_df.append({"Block": name, "Cash(T)": cash, "Gold(T)": gold, "Oil(kkk)":oil, "Ore(kkk)":ore,
                                "Uranium(kk)":uranium,  "Dia(kk)":dia}, ignore_index=True)
#blocs_df['Gold'] = blocs_df['Gold'].apply(lambda x: '{:.0f}'.format(x))
#blocs_df.sort_values(by="Gold(T)", inplace=True, ascending=False)

with pd.ExcelWriter(f'resources_bloc_{month}_{day}.xlsx', engine='openpyxl') as writer:
    
    blocs_df.to_excel(writer, sheet_name='Sheet1', index=True)
    # Openpyxl ile Excel dosyasını düzenleme
    workbook = writer.book
    worksheet = writer.sheets['Sheet1']
    
    # Sütun aralığını belirt
    column_range = f"A1:{chr(65+len(blocs_df.columns))}{len(blocs_df)+1}"
    print(column_range)
    # Tablo oluşturma
    worksheet.auto_filter.ref = column_range 
```
